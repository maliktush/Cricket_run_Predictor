```python
def predict2(numls,numos,numrf,numlf,numm,cntry,opp,ovr):
    
    sumweighted=numls*lsavg + numos*osavg + numlf*lfavg + numrf*rfavg + numm*mavg
    no_bowlers=numls+numos+numrf+numlf+numm
    pred=sumweighted/no_bowlers
    #print(pred)
    cntryavg=0
    oppavg=0
    ovravg=0
    if(cntry=="Home"):
        cntryavg=indavg
    if(cntry=="Australia"):
        cntryavg=ausavg
    if(cntry=="New Zealand"):
        cntryavg=nzavg
    if(cntry=="South Africa"):
        cntryavg=saavg
    if(cntry=="England"):
        cntryavg=engavg
    if(cntry=="West Indies"):
        cntryavg=wiavg
    if(cntry=="SriLanka"):
        cntryavg=slavg
    if(cntry=="Bangladesh"):
        cntryavg=bdavg
    
    #print(pred2)
    if ovr<10:
        ovravg=avg0
    if ovr<20 and ovr>10:
        ovravg=avg1
    if ovr<50 and ovr>20:
        ovravg=avg2
    if ovr<75 and ovr>50:
        ovravg=avg5
    if ovr>75:
        ovravg=avg7
   
    if(opp=="Australia"):
        oppavg=againstausavg
    if(opp=="New Zealand"):
        oppavg=againstnzavg
    if(opp=="South Africa"):
        oppavg=againstsaavg
    if(opp=="England"):
        oppavg=againstengavg
    if(opp=="West Indies"):
        oppavg=againstwiavg
    if(opp=="SriLanka"):
        oppavg=againstslavg
    if(opp=="Bangladesh"):
        oppavg=againstbdavg
    predfinal=(pred+cntryavg+ovravg+oppavg)/4
    return predfinal
```
